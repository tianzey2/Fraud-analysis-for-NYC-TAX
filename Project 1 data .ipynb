```python
mean=df.groupby('TAXCLASS')['STORIES'].mean()
print(mean)
```

```python
df['STORIES']=df['STORIES'].fillna(value=0)
for index in mean.index:
    df.loc[(df['STORIES']==0)&(df['TAXCLASS']==index),'STORIES']=mean[index]
```

```python
df['AVLAND'].replace('NaN',0)
notzero=df[df['AVLAND']!=0]
meanavl=notzero.groupby('TAXCLASS')['AVLAND'].mean()
meanavl
```

```python
missing_avland=df[df['AVLAND'].isnull()]
len(missing_avland)
```

```python
df['FULLVAL'].replace('NaN',0)
notzero=df[df['FULLVAL']!=0]
meanful=notzero.groupby('TAXCLASS')['FULLVAL'].mean()
meanful
```

```python
missing_fullval=df[df['FULLVAL'].isnull()]
len(missing_fullval)
```

```python
df['AVTOT'].replace('NaN',0)
notzero=df[df['AVTOT']!=0]
meanavt=notzero.groupby('TAXCLASS')['AVTOT'].mean()
meanavt
```

```python
missing_avtot=df[df['AVTOT'].isnull()]
len(missing_avtot)
```

```python
%%time
for i in meanful.index:
    df.loc[(df['FULLVAL']==0)&(df['TAXCLASS']==i),'FULLVAL']=meanful[i]
    df.loc[(df['AVLAND']==0)&(df['TAXCLASS']==i),'AVLAND']=meanavl[i]
    df.loc[(df['AVTOT']==0)&(df['TAXCLASS']==i),'AVTOT']=meanavt[i]
```

```python
df.describe()
```

```python
missing_ltf=df['LTFRONT'].isnull()
len(missing_ltf)
```

```python
df.loc[df['LTFRONT']==0,'LTFRONT']=np.nan
df.loc[df['LTDEPTH']==0,'LTDEPTH']=np.nan
df.loc[df['BLDFRONT']==0,'BLDFRONT']=np.nan
df.loc[df['BLDDEPTH']==0,'BLDDEPTH']=np.nan
```

```python
# calculate mean 
meanltf=df.groupby(df['TAXCLASS'])['LTFRONT'].mean()
meanltd=df.groupby(df['TAXCLASS'])['LTDEPTH'].mean()
meanbldf=df.groupby(df['TAXCLASS'])['BLDFRONT'].mean()
meanbldd=df.groupby(df['TAXCLASS'])['BLDDEPTH'].mean()
```

```python
# change values
for i in meanltf.index:
    df.loc[(df['LTFRONT'].isnull())&(df['TAXCLASS']==i),'LTFRONT']=meanltf[i]    
    df.loc[(df['LTDEPTH'].isnull())&(df['TAXCLASS']==i),'LTDEPTH']=meanltd[i]
    df.loc[(df['BLDFRONT'].isnull())&(df['TAXCLASS']==i),'BLDFRONT']=meanbldf[i]
    df.loc[(df['BLDDEPTH'].isnull())&(df['TAXCLASS']==i),'BLDDEPTH']=meanbldd[i]
```

```python
df2=pd.read_csv('cleandata.csv',index_col=0)
```

```python
df2.head()
```

```python
df2.describe()
```

```python
df2["LOTAREA"]=df2.LTFRONT*df2.LTDEPTH
df2["BLDAREA"]=df2.BLDFRONT*df2.BLDDEPTH
df2["BLDVOL"]=df2.BLDAREA*df2.STORIES
```

```python
df2.describe()
```

```python
df2["r1"]=df2.FULLVAL/df2.LOTAREA
df2["r2"]=df2.FULLVAL/df2.BLDAREA
df2["r3"]=df2.FULLVAL/df2.BLDVOL
df2["r4"]=df2.AVLAND/df2.LOTAREA
df2["r5"]=df2.AVLAND/df2.BLDAREA
df2["r6"]=df2.AVLAND/df2.BLDVOL
df2["r7"]=df2.AVTOT/df2.LOTAREA
df2["r8"]=df2.AVTOT/df2.BLDAREA
df2["r9"]=df2.AVTOT/df2.BLDVOL
```

```python
df2.describe()
```

```python
df2['ZIP']=df2['ZIP'].astype(str)
df2['ZIP'].describe()
```

```python
df2['zip3']=df2['ZIP'].str[:3]
```

```python
df2.head()
```

```python
vars=['r1','r2','r3','r4','r5','r6','r7','r8','r9']
```

```python
meanzip5=df2.groupby('ZIP')[vars].mean()
meanzip3=df2.groupby('zip3')[vars].mean()
meantaxclass=df2.groupby('TAXCLASS')[vars].mean()
meanboro=df2.groupby('B')[vars].mean()
```

```python
df2=df2.join(meanzip5,on='ZIP',rsuffix='_zip5')
df2=df2.join(meanzip3,on='zip3',rsuffix='_zip3')
df2=df2.join(meantaxclass,on='TAXCLASS',rsuffix='_taxclass')
df2=df2.join(meanboro,on='B',rsuffix='_boro')
```

```python
resuffix=['_zip5','_zip3','_taxclass','_boro']
for i in vars:
    for k in resuffix:
        df2[str(i)+k]=df2[i]/df2[str(i)+k]
```

```python
df2.describe()
```

```python
result=df2.drop(columns=['FULLVAL','AVLAND','STORIES','LTFRONT','BLDFRONT','BLDDEPTH','ZIP','zip3','TAXCLASS','B','AVTOT','LTDEPTH','LOTAREA'
                        ,'BLDAREA','BLDVOL'])
```
